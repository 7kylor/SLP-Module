```python

class NoteCatalog:
    def __init__(self):
        self.note_details = {}
    
    def add_note(self, title, Id):
        self.note_details[Id] = title
        print ((f"Note {title} with Id {Id} is added to the notes"))
        
    def get_note_by_Id(self, Id):
        if Id in self.note_details:
            return self.note_details[Id]
        else:
            return "note with + {Id} + not found"

def main():
    notes = NoteCatalog()
    
    while True:
        print("1. Add New note")
        print("2. Get a note by id")
        print("3. Exit")
        choice =  (input("Enter your choice: "))
        if choice ==   "1":
            title = input("Enter the note title: ")
            Id = input("Enter the note id: ")
            notes.add_note(title, Id)
            
        elif choice ==   "2":
            Id =  ((input("Enter Note id")))
            title =  notes.get_note_by_Id(Id)
            print(f"the note your looking for is {title}")  
        elif choice ==  "3":
            break
        else:
            print("Invalid choice. Try again.")
            continue
        
        
if __name__ == "__main__":
    main()
        
```
