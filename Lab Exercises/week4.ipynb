```python
# -*- coding: week 4-*-


# Exercise 1:
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        
    def display(self):
        print("Name: ", self.name, ", Salary: ", self.salary)
        
        
object1 = Employee("Abdularahaman", 2000)
object2 = Employee("Taher", 5000)

object1.display()
object2.display()


# Exercise 2:


class Students:
    
    def __init__(self, name, grade, specialization):
        self.name = name
        self.grade = grade
        self.specialization = specialization
    
    
    def display(self):
        print("Name: ", self.name, ", Grade: ", self.grade, ", Specialization: ", self.specialization)
    
    
object1 = Students("Abdularahaman", 98, "Computer Science")
object2 = Students("Taher", 97, "Software Engineering")

object1.display()
object2.display()


#  Exercise 3 form the slides 

class StudentsDB:
    def __init__(self):
        self.students = {}
    
    def add_student(self, name, id):
        self.students[id] = name
        
    def get_student(self, name , id):
      if id in self.students:
        print("Student name: ", self.students[id])
      else:
        print("Student not found")
      
def main():
    info = StudentsDB()  
    while True:
        print("1. Add student" '\n' "2. Get student" '\n' "3. Exit")
        choice = int(input("Do you want to add or get student?"))
        
        if choice == "1":
            name = input("Enter student name: ")
            id = input("Enter student id: ")
            info.add_student(name, id)
        elif choice == "2":
            id = input("Enter student id: ")
            info.get_student(id)
        elif choice == "3":
            print("Exiting program")
            break
        else:
            print("Invalid choice")
       
        
if __name__ == "__main__":
    main()




```
