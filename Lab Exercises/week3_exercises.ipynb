```python
# -*- coding: week 3 exercises -*-

import random 
import math
from collections import Counter 


 #1
friends = ['Ali', 'Abdurahamn', 'Salim', 'Ammar', 'Al Baraa']
friends.append('Majid')
friends.insert(1, 'Omar')
friends.remove('Salim')
friends.extend(['Amjed', 'Omar'])
friends.sort(reverse=True)
# friends.reverse()
print("\n".join(friend for friend in friends)  + '\n')


top_favorite_food = ['Pizza', 'Chocolate', 'Pasta']
print("\n".join(f'{food} is my one of top favorite food' for food in top_favorite_food))
    
numbers = [3,5,7]
for number in numbers:
    print(f'The square root of {number} is {math.sqrt(number)}' + '\n')
 
    
# #2
def generate_weather_data():
    weather = ['Sunny', 'Rainy', 'Snowy', 'Stormy', 'Foggy']
    return [(f'2024-3-{day}', random.choice(weather)) for day in range(1, 31)]

for date, weather in generate_weather_data():
    print(f'On {date}, the weather was {weather}.')

#3 
def generate_Weather_data():
    weather = ['Sunny', 'Rainy', 'Snowy', 'Stormy', 'Foggy']
    date = [f'2024-3-{day}' for day in range(1, 31)]
    return [(date, random.choice(weather)) for date in date]

for date, weather in generate_Weather_data():
    print(f'On {date}, the weather was {weather}.')

#4
def generate_weather_data():
    weather = ['Sunny', 'Rainy', 'Snowy', 'Stormy', 'Foggy']
    date = [f'2024-3-{day}' for day in range(1, 31)]
    temperature = [random.randint(-10, 40) for _ in range(30)]
    rainfall = [random.randint(0, 100) for _ in range(30)]
    return [(date, random.choice(weather), temp, rain) for date, temp, rain in zip(date, temperature, rainfall)]
for date, weather, temp, rain in generate_weather_data():
    print(f'On {date}, the weather was {weather} with a temperature of {temp}C and {rain}% chance of rain.')
 
#5
weather_data = generate_weather_data()
temperature = [temp for date, weather, temp, rain in weather_data]
avg_temp = sum(temperature) / len(temperature)
max_temp = max(temperature)
most_common_temp = Counter(temperature).most_common(1)
day_with_max_temp =  max(weather_data, key=lambda x: x[2])  # x[2] is the temperature
for date, weather, temp, rain in weather_data:
    print(f'On {date}, the weather was {weather} with a temperature of {temp}C and {rain}% chance of rain.')
    print(f'The average temperature was {avg_temp:.2f}C.')
    print(f'The day with the highest temperature was {day_with_max_temp[0]}.')
    print(f'The most common temperature was {most_common_temp[0][0]}C.')

 #6  
def generate_sales_data():
    date = [f'2024-3-{day}' for day in range(1, 31)]
    sales = [random.randint(100, 1000) for _ in range(30)]
    transactions = [random.randint(10, 100) for _ in range(30)]
    return [(date, sale, transaction) for date, sale, transaction in zip(date, sales, transactions)]

sales_data = generate_sales_data()
daily_sales = [sale for date, sale, transaction in sales_data]
day_with_max_sales = max(sales_data, key=lambda x: x[1])  # x[1] is the sale amount

avg_sales = sum(daily_sales) / len(daily_sales)
most_common_sales = Counter(daily_sales).most_common(1)

for date, sale, transaction in sales_data:
    print(f'On {date}, there were {transaction} transactions with a total of ${sale} in sales.')

print(f'The average daily sales was ${avg_sales:.2f}.')
print(f'The day with the highest sales was {day_with_max_sales[0]}.')
print(f'The most common sales amount was ${most_common_sales[0][0]}.')


#7
def add_contact(contacts, name, number):
    contacts[name] = number
    return contacts

def update_contact(contacts, name, number):
    if name in contacts:
        contacts[name] = number
    else:
        print(f'{name} not found in contacts.')
    return contacts

def retrieve_contact(contacts, name):
    if name in contacts:
        return contacts[name]
    else:
        return f'{name} not found in contacts.'
    
def display_menu():
    print('1. Add contact')
    print('2. Update contact')
    print('3. Retrieve contact')
    print('4. Exit')
    return input('Enter your choice: ')

contacts = {}
while True:
    choice = display_menu()
    if choice == '1':
        name = input('Enter name: ')
        number = input('Enter number: ')
        contacts = add_contact(contacts, name, number)
    elif choice == '2':
        name = input('Enter name: ')
        number = input('Enter number: ')
        contacts = update_contact(contacts, name, number)
    elif choice == '3':
        name = input('Enter name: ')
        print(retrieve_contact(contacts, name))
    elif choice == '4':
        break
    else:
        print('Invalid choice. Try again.')
        
```
