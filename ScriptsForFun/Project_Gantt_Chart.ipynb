```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define task data
tasks = [
    {"Task": "Project Start", "Start": "2024-06-02", "Finish": "2024-06-06", "Duration": 5},
    {"Task": "Requirement Gathering", "Start": "2024-06-16", "Finish": "2024-06-20", "Duration": 5},
    {"Task": "UI/UX Design", "Start": "2024-06-23", "Finish": "2024-07-15", "Duration": 17},
    {"Task": "Prototyping", "Start": "2024-07-16", "Finish": "2024-08-05", "Duration": 15},
    {"Task": "Backend Development", "Start": "2024-07-07", "Finish": "2024-09-09", "Duration": 47},
    {"Task": "Frontend Development (Android)", "Start": "2024-08-06", "Finish": "2024-09-29", "Duration": 39},
    {"Task": "Frontend Development (iOS)", "Start": "2024-08-08", "Finish": "2024-09-29", "Duration": 39},
    {"Task": "API Development and Integration", "Start": "2024-10-01", "Finish": "2024-10-31", "Duration": 24},
    {"Task": "Unit Testing", "Start": "2024-11-03", "Finish": "2024-11-11", "Duration": 7},
    {"Task": "Integration Testing", "Start": "2024-11-14", "Finish": "2024-11-20", "Duration": 9},
    {"Task": "System Testing", "Start": "2024-11-25", "Finish": "2024-12-01", "Duration": 9},
    {"Task": "User Acceptance Testing (UAT)", "Start": "2024-12-07", "Finish": "2024-12-14", "Duration": 4},
    {"Task": "App Store Submission", "Start": "2024-12-17", "Finish": "2024-12-21", "Duration": 3},
    {"Task": "Deployment to Production", "Start": "2024-12-22", "Finish": "2024-12-26", "Duration": 5},
    {"Task": "Documentation", "Start": "2024-12-28", "Finish": "2025-01-03", "Duration": 4},
    {"Task": "Project Review", "Start": "2025-01-04", "Finish": "2025-01-10", "Duration": 5},
]

# Create a DataFrame
df = pd.DataFrame(tasks)

# Convert dates to datetime
df['Start'] = pd.to_datetime(df['Start'])
df['Finish'] = pd.to_datetime(df['Finish'])

# Plotting the Gantt chart
fig, ax = plt.subplots(figsize=(12, 8))

# Bars for each task
for idx, task in df.iterrows():
    ax.barh(task['Task'], (task['Finish'] - task['Start']).days, left=task['Start'], height=0.5, color='skyblue', edgecolor='black')

# Formatting the dates on x-axis
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.xticks(rotation=45)

# Labels and title
plt.xlabel('Date')
plt.ylabel('Task')
plt.title('Project Gantt Chart')

# Show grid
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Save the plot as an image
plt.tight_layout()
plt.savefig('Project_Gantt_Chart.png')

plt.show()
```
