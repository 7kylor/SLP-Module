```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the project tasks and their dependencies
tasks = {
    "A": {"name": "Project Start", "duration": 5},
    "B": {"name": "Requirement Gathering", "duration": 5},
    "C": {"name": "UI/UX Design", "duration": 17},
    "D": {"name": "Prototyping", "duration": 15},
    "E": {"name": "Backend Development", "duration": 24},
    "F": {"name": "Frontend Development (Android)", "duration": 39},
    "G": {"name": "Frontend Development (iOS)", "duration": 39},
    "H": {"name": "API Development and Integration", "duration": 24},
    "I": {"name": "Unit Testing", "duration": 7},
    "J": {"name": "Integration Testing", "duration": 9},
    "K": {"name": "System Testing", "duration": 9},
    "L": {"name": "User Acceptance Testing (UAT)", "duration": 4},
    "M": {"name": "App Store Submission", "duration": 3},
    "N": {"name": "Deployment to Production", "duration": 5},
    "O": {"name": "Documentation", "duration": 4},
    "P": {"name": "Project Review", "duration": 5},
}

dependencies = [
    ("A", "B"),  # Project Start -> Requirement Gathering
    ("B", "C"),  # Requirement Gathering -> UI/UX Design
    ("C", "D"),  # UI/UX Design -> Prototyping
    ("D", "E"),  # Prototyping -> Backend Development
    ("E", "F"),  # Backend Development -> Frontend Development (Android)
    ("E", "G"),  # Backend Development -> Frontend Development (iOS)
    ("E", "H"),  # Backend Development -> API Development and Integration
    ("F", "I"),  # Frontend Development (Android) -> Unit Testing
    ("G", "I"),  # Frontend Development (iOS) -> Unit Testing
    ("H", "I"),  # API Development and Integration -> Unit Testing
    ("I", "J"),  # Unit Testing -> Integration Testing
    ("J", "K"),  # Integration Testing -> System Testing
    ("K", "L"),  # System Testing -> User Acceptance Testing (UAT)
    ("L", "M"),  # User Acceptance Testing (UAT) -> App Store Submission
    ("M", "N"),  # App Store Submission -> Deployment to Production
    ("N", "O"),  # Deployment to Production -> Documentation
    ("O", "P"),  # Documentation -> Project Review
]

# Create a directed graph
G = nx.DiGraph()

# Add nodes with task durations
for task, info in tasks.items():
    G.add_node(task, label=f"{task}\n{info['name']}\n({info['duration']} days)")

# Add edges (dependencies)
G.add_edges_from(dependencies)

# Define node positions
pos = nx.spring_layout(G, seed=20)  # Positions for all nodes

# Draw the network diagram
plt.figure(figsize=(25, 19))  # Increase the figure size
nx.draw(G, pos, with_labels=True, node_size=9000, node_color='skyblue', font_size=9, font_weight='bold', labels=nx.get_node_attributes(G, 'label'))
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f"{tasks[v]['duration']} days" for u, v in G.edges}, font_color='red')

plt.title('Critical Path Method (CPM) Network Diagram')
plt.savefig('CPM_Network_Diagram.png')
plt.show()
```
