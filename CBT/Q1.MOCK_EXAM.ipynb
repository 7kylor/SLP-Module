```python
# Q1.  MOCK EXAM LAB EXAM-1

class MusicLibrary:
    def __init__(self):
        self.song_records = {}

    def add_song(self, song_id, artist_name):
        self.song_records[song_id] = artist_name
        print(f"Song with ID {song_id} by {artist_name} added successfully.")

    def get_song(self, song_id):
        if song_id in self.song_records:
            return self.song_records[song_id]
        else:
            return "Song not found."
        
        
def main():
    music_library = MusicLibrary()
    while True:
        print("Menu:")
        print("a. Add Song")
        print("b. Get Song")
        print("c. Exit")
        choice = input("Enter your choice: ")
        if choice == "a":
            song_id = input("Enter song ID: ")
            artist_name = input("Enter artist name: ")
            music_library.add_song(song_id, artist_name)
        elif choice == "b":
            song_id = input("Enter song ID: ")
            artist_name = music_library.get_song(song_id)
            print(f"Artist name: {artist_name}")
        elif choice == "c":
            break
        else:
            print("Invalid choice. Please try again.")
            
if __name__ == "__main__":
    main()
```
